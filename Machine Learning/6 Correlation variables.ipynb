```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as matplotlib
import seaborn as sn
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
FILE_TRAIN = 'data/pp-training.csv'
FILE_TEST = 'data/pp-testing.csv'
```

```python
df_test = pd.read_csv(FILE_TEST, low_memory=False)
```

```python
df_train = pd.read_csv(FILE_TRAIN, low_memory=False)
```

```python
corrMatt = df_train.corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(30,30)
sn.heatmap(corrMatt, mask=mask,vmax=.8, square=True)
```
