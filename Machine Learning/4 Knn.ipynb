```python
import pandas as pd
import numpy as np
```

```python
FILE_TRAIN = 'data/pp-training.csv'
FILE_TEST = 'data/pp-testing.csv'
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(FILE_TRAIN, low_memory=False)
```

```python
allcols = df.columns
alldata = df
X_train, X_test, y_train, y_test = train_test_split(alldata.drop(['price'], axis=1),
                                                    alldata.price, test_size=0.1)
```

```python

regr = KNeighborsRegressor(n_neighbors=8, algorithm='kd_tree', metric='manhattan', n_jobs=-1)
regr.fit(X_train, y_train)
regr.score(X_test, y_test)
```

```python
regr = KNeighborsRegressor(metric='euclidean' ,n_neighbors=20, algorithm='kd_tree', weights='distance', n_jobs=-1)
regr.fit(X_train, y_train)
regr.score(X_test, y_test)
```

```python
cols = list(X_train)
```

```python
df = pd.read_csv(FILE_TEST, low_memory=False, usecols=cols)
df = df[cols]
predicted_2 = regr.predict(df)
```

```python
ids = pd.read_csv('data/properati_dataset_testing_noprice.csv')
ids = ids[['id']]
df2 = pd.DataFrame(predicted_2)
df3 = ids.join(df2)
df3.columns = ['id','price_usd']
df3.to_csv('data/prediccion-kaggle-Knn.csv', index=False)
df3 = df3.set_index('id')
```
