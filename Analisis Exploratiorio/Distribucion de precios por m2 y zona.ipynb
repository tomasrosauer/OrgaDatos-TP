```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('propiedades-clean.csv')
df.tail()
```

```python
#CABA properties
caba = df.loc[df.state_name.str.contains('Capital Federal'), :]

#GBA Norte properties
gba_n = df.loc[df.state_name.str.contains('Bs.As. G.B.A. Zona Norte'), :]

#GBA Oeste properties
gba_o = df.loc[df.state_name.str.contains('Bs.As. G.B.A. Zona Oeste'), :]

#GBA Sur properties
gba_s = df.loc[df.state_name.str.contains('Bs.As. G.B.A. Zona Sur'), :]
```

### How are the prices distributed in CABA?

```python
no_outliers_caba = caba[np.abs(caba.price_usd_per_m2 - caba.price_usd_per_m2.mean()) <= (3 * caba.price_usd_per_m2.std())]

fig = plt.figure(figsize=(20, 10))
ax = plt.subplot(111)
ax.spines['left'].set_color('#1a1a1a')
ax.spines['bottom'].set_color('#1a1a1a')

ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left()

plt.xticks(fontsize=14, color='#1a1a1a')  
plt.yticks(fontsize=14, color='#1a1a1a')

plt.title('Distribuciรณn de precio por m2 en CABA', fontsize=30)
plt.xlabel('Precio (usd)', fontsize=20)
plt.ylabel('Frecuencia', fontsize=20)

plt.hist(no_outliers_caba.price_usd_per_m2, bins=100, color='grey');
```
