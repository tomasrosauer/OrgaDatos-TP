```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('default') # haciendo los graficos un poco mas bonitos xD
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
propiedades = pd.read_csv('propiedades-clean.csv', low_memory=False)
```

```python
grouped = propiedades.loc[:,['place_name','price_aprox_usd']].groupby('place_name').\
    agg([np.mean,np.size])
top20 = grouped.loc[grouped[('price_aprox_usd','size')]>1000,:].sort_values(('price_aprox_usd','mean'), ascending=False)    
top20.plot(kind='bar',figsize=(60,20))
```

```python
#azul, tamaño promedio del terreno,naranja, superficie total de todos los terrenos
grouped = propiedades.loc[:,['place_name','surface_total_in_m2']].groupby('place_name').\
    agg([np.mean,np.size])
top20 = grouped.loc[grouped[('surface_total_in_m2','size')]>1000,:].sort_values(('surface_total_in_m2','mean'), ascending=False)    
top20.plot(kind='bar',figsize=(60,20))
```
